```python
import numpy
import cv2
import matplotlib.pyplot as plt
from tqdm import tqdm
import os
import asyncio
import time
```

```python
folder = '/home/sghosal/Project/Image_Inpainting//testSet_resize/'
```

```python
def background(f):
    def wrapped(*args, **kwargs):
        return asyncio.get_event_loop().run_in_executor(None, f, *args, **kwargs)

    return wrapped

@background
def crop_image(file,i):
    img = cv2.imread(folder + '/' + file)
    plt.imsave(f'/home/sghosal/Project/Image_Inpainting/data_places_gt/{i}.jpg', img)
#     img[96:96 + 64, 96:96 + 64, :] = 0
#     plt.imsave(f'./data_places/{i}.jpg', img)

files = os.listdir(folder)
for i in tqdm(range(41000)):
    crop_image(files[i], i)

print('loop finished')
```
